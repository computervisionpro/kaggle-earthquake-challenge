# Scaling numeric data

```python
from sklearn.preprocessing import StandardScaler, RobustScaler

# Columns with high outliers
other_cols = list(set(important_cols)-set(['feature_3', 'feature_4', 'feature_5', 'feature_14']))
print(other_cols)
# # Normal colunms
# normal_cols = ['age', 'education-num', 'hours-per-week']

# Apply RobustScaler to outliers
robust = RobustScaler()
X_train[other_cols] = robust.fit_transform(X_train[other_cols])
X_test[other_cols] = robust.transform(X_test[other_cols])

# log scaling
for col in ['feature_3', 'feature_4', 'feature_5']:
    X_train[col] = np.log1p(X_train[col])
    X_test[col] = np.log1p(X_test[col])
    
# Apply StandardScaler to the rest
normal_col = ['feature_14']
std = StandardScaler()
X_train[normal_col] = std.fit_transform(X_train[normal_col])
X_test[normal_col] = std.transform(X_test[normal_col])
```

#### We are Scaling the data after splitting, as Test data would be unknown to the model and then apply the scaling on test data using transform method

```python
import xgboost as xgb
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from skorch import NeuralNetRegressor
from model_pt import MagnitudeRegressor

torch_regressor = NeuralNetRegressor(
    MagnitudeRegressor,
    module__input_dim=16,       
    criterion=nn.SmoothL1Loss,   
    optimizer=optim.AdamW,       
    max_epochs=300,               
    lr=0.001,                   
    batch_size=32,
    iterator_train__shuffle=True,
    train_split=None,  # IMPORTANT for stacking
    device='cuda',
    verbose=0
)
```

```python
import lightgbm as lgb
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

lgb_model = lgb.LGBMRegressor(
    n_estimators=1500,
    learning_rate=0.03,
    num_leaves=20,          # controls complexity
    max_depth=-1,           # no limit
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.1,          # L1 regularization
    reg_lambda=1.0,         # L2 regularization
    random_state=42,
    n_jobs=-1
)
```

```python
# Approach A: sklearn StackingRegressor

from sklearn.ensemble import GradientBoostingRegressor, StackingRegressor
from sklearn.linear_model import Ridge, ElasticNet, RidgeCV
from sklearn.pipeline import Pipeline




# Define base learners
base_learners = [
    ('rf', RandomForestRegressor(n_estimators=800, random_state=42, n_jobs=-1)),
    ('xgb', xgb_model),
    ('dnn', torch_regressor), # ('mlp', mlp_model),
    ('lightgb', lgb_model),
    ('svr', SVR(C=15, epsilon=0.1, gamma='scale')),
    
]

# Meta-learner
meta_pipeline = Pipeline([
    ('scaler_meta', RobustScaler()),
    ('ridge', Ridge(alpha=1.0))
    # ('ridge', RidgeCV(alphas=[5, 7, 10], cv=2))
                ])


# pipeline
stack = StackingRegressor(
    estimators=base_learners,
    final_estimator=meta_pipeline,
    cv=5, #3
    n_jobs=-1,
    passthrough=False 
)
```

#### We are using five models so we can get better aggregation of prediction. We used Robust Scaler before ridge as it is sensitive to scaling and to provide better accuracy.

```python
# Convert data
X_train = X_train.values.astype(np.float32)
y_train = y_train.values.astype(np.float32)
```

```python
# contribution
meta_model = stack.final_estimator_.named_steps['ridge']
coefs = meta_model.coef_

model_names = [name for name, _ in stack.estimators]

# Get OOF predictions used by meta model
X_meta = stack.transform(X_train)

stds = X_meta.std(axis=0)
true_contrib = coefs * stds

for name, val in zip(model_names, true_contrib):
    print(f"{name}: {val:.4f}")
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import StratifiedKFold

n_bins = 5

try:
    y_binned = pd.qcut(y_train, q=n_bins, labels=False, duplicates='drop')
except ValueError:
    # fallback to equal-width bins if qcut fails
    y_binned = pd.cut(y_train, bins=n_bins, labels=False)

y_binned = y_binned.astype(int)

# Check bin distribution
bin_counts = np.bincount(y_binned)
print("Bin counts:", bin_counts)
```

```python
# Inverse frequency weights
weights_per_bin = 1 / bin_counts

# Normalize
weights_per_bin = weights_per_bin / np.mean(weights_per_bin)

# Assign weight to each sample
sample_weights = weights_per_bin[y_binned]
sample_weights
print(len(sample_weights))
```
